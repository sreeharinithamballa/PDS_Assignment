```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/StudentsPerformance (1).csv')
```

```python
df
```

Visualizations

Bar plot

```python

# Visualization 1 - Bar plot of average writing score by gender
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='writing score', data=df, errorbar=None, palette='rocket',hue='gender', legend=False)
plt.title('Average writing Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average writing Score')
plt.savefig('/visualization_1.png')
plt.show()



```

Box Plot

```python
# Visualization 2 - Box Plot of Writing Scores Grouped by Test Preparation Course
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='test preparation course', y='writing score', palette='pastel', hue='test preparation course', legend=False )
plt.title('Writing Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Writing Score')
plt.savefig('/visualization_2.png')
plt.show()
```

Scatter Plot

```python
# Visualization 3 - Scatter plot of math score vs reading score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', data=df, hue='gender')
plt.title('Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.grid(True)
plt.savefig('/visualization_3.png')
plt.show()
```

Histogram Plot

```python
# Visualization 4 - Histogram of math score
plt.figure(figsize=(8, 6))
sns.histplot(df['math score'], kde=False, stat='count', bins=20)
plt.title('Distribution of math Score')
plt.xlabel('Math Score')
plt.ylabel('count')
plt.grid(True)
plt.savefig('/visualization_4.png')
plt.show()
```

Heat Map

```python
# Visualization 5 - Correlation Heatmap of Test Scores
plt.figure(figsize=(8,6))
correlation_matrix = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Heatmap of Test Scores')
plt.savefig('/visualization_5.png')
plt.show()
```
