```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/rawData.csv')
```

```python
df
```

```python
df['Frailty'].replace({'N': 0, 'Y': 1},inplace=True)
```

```python
#Performing the 2 sample t-test
from scipy import stats
t2_test = stats.ttest_ind(df['Weight'],df['Frailty'])
print('analysis:',t2_test)
```

Visualizations

```python
# Visualization 1 - Bar plot of Average Grip Strength by Frailty
avg = df.groupby('Frailty')['Grip strength'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.bar(avg['Frailty'], avg['Grip strength'], color=['lightblue', 'lightgreen'])
plt.xlabel('Frailty')
plt.ylabel('Average Grip Strength')
plt.title('Average Grip Strength by Frailty')
plt.show()
```

```python
# Visualization 2 - Scatterplot of Age vs Grip Strength
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Grip strength'], color='green')
plt.xlabel('Age')
plt.ylabel('Grip Strength')
plt.title('Scatterplot of Age vs Grip Strength')
plt.show()
```

```python
# Visualization 3 - Box plot of Age
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Age')
plt.ylabel('Age (years)')
plt.title('Box plot of Age')
plt.show()
```
